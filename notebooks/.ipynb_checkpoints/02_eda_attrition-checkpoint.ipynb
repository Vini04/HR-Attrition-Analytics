```python
# Necessary Import Statements 
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
import sys
sys.executable
```

```python
# Load Data
```

```python
df = pd.read_csv("../data/processed/hr_attrition_fe.csv")
```

```python
# Overall Attrition Rate
overall_attrition = df['AttritionFlag'].mean()
```

```python
overall_attrition
```

```python
# Overall employee attrition rate is ~16%, indicating moderate but non-negligible workforce churn.
```

```python
# ATTRITION BY DEPARTMENT : Which departments are bleeding people?
```

```python
dept_attrition = (
    df.groupby('Department')['AttritionFlag']
      .mean()
      .sort_values(ascending=False)
)

dept_attrition
```

```python
# ATTRITION BY DEPARTMENT : Plot
```

```python
dept_attrition.plot(kind='bar')
plt.title('Attrition Rate by Department')
plt.ylabel('Attrition Rate')
plt.show()
```

```python
# Above : Sales shows the highest attrition, likely due to performance pressure and variable compensation structure.
```

```python
# ATTRITION BY JOB ROLE : Which roles are high-risk?
```

```python
role_attrition = (
    df.groupby('JobRole')['AttritionFlag']
      .mean()
      .sort_values(ascending=False)
)

role_attrition
```

```python
# ATTRITION BY JOB ROLE : Plot
```

```python
role_attrition.plot(kind='bar', figsize=(10,5))
plt.title('Attrition Rate by Job Role')
plt.ylabel('Attrition Rate')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Above : Frontline and junior roles show significantly higher attrition compared to senior or managerial roles.
```

```python
# OVERTIME VS ATTRITION : Is overtime causing churn?
```

```python
overtime_attrition = (
    df.groupby('OverTime')['AttritionFlag']
      .mean()
)

overtime_attrition
```

```python
# OVERTIME VS ATTRITION : Plot
```

```python
overtime_attrition.plot(kind='bar')
plt.title('Attrition Rate by Overtime')
plt.ylabel('Attrition Rate')
plt.show()
```

```python
# Above : Employees working overtime exhibit nearly 3 higher attrition, making overtime one of the strongest predictors of employee exit.
```

```python
# SALARY BAND VS ATTRITION : Are we underpaying people who leave?
```

```python
salary_attrition = (
    df.groupby('SalaryBand')['AttritionFlag']
      .mean()
      .sort_values(ascending=False)
)

salary_attrition
```

```python
# SALARY BAND VS ATTRITION : Plot
```

```python
salary_attrition.plot(kind='bar')
plt.title('Attrition Rate by Salary Band')
plt.ylabel('Attrition Rate')
plt.show()
```

```python
# Above : Employees in the low salary band experience significantly higher attrition, suggesting compensation plays a critical retention role.
```

```python
# AGE GROUP VS ATTRITION : Who leaves more  early career or senior employees?
```

```python
age_attrition = (
    df.groupby('AgeGroup')['AttritionFlag']
      .mean()
)

age_attrition
```

```python
# AGE GROUP VS ATTRITION : Plot
```

```python
age_attrition.plot(kind='bar')
plt.title('Attrition Rate by Age Group')
plt.ylabel('Attrition Rate')
plt.show()
```

```python
# Above : Early-career employees (1825, 2635) show higher attrition compared to senior age groups, indicating early-tenure instability.
```

```python
# JOB SATISFACTION VS ATTRITION : Does dissatisfaction actually cause exits?
```

```python
js_attrition = (
    df.groupby('JobSatisfactionLevel')['AttritionFlag']
      .mean()
      .reindex(['Low', 'Medium', 'High', 'Very High'])
)

js_attrition
```

```python
# JOB SATISFACTION VS ATTRITION : Plot
```

```python
js_attrition.plot(kind='bar')
plt.title('Attrition Rate by Job Satisfaction')
plt.ylabel('Attrition Rate')
plt.show()
```

```python
# Above : Attrition decreases monotonically with higher job satisfaction, validating satisfaction as a key retention lever.
```

```python
# TENURE BUCKET VS ATTRITION : When do employees quit?
```

```python
df.groupby('TenureBucket')['AttritionFlag'].mean()
```

```python
# TENURE BUCKET VS ATTRITION : Plot
```

```python
tenure_attrition = (
    df.groupby('TenureBucket')['AttritionFlag']
      .mean()
      .reindex(['0-2', '3-5', '6-10', '10+'])
)

tenure_attrition.plot(kind='bar')
plt.title('Attrition Rate by Tenure')
plt.ylabel('Attrition Rate')
plt.show()
```

```python
# Above : 
# Employees are most likely to leave within their first two years, suggesting weaknesses in onboarding, early role alignment, or initial engagement.
```

```python
# Attrition is front-loaded. Nearly one-third of employees leave within the first two years, after which attrition steadily declines. 
# This points to onboarding, early workload, and manager alignment issues rather than long-term dissatisfaction.
```
